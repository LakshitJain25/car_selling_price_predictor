Lets deal with torque now

**what does torque signify?**

Applied to internal combustion engines or electric motors, torque indicates the force to which the drive shaft is subjected.

Torque is expressed in pound-feet (lb-ft) or newton-meters (Nm).

The interaction of torque and engine speed (rpm) determines the engine power.

Since Rpm remains constant unit in the torque column we will extract it




```python
import re
def extractRPM(torque):
  torque = re.sub(",","",torque)
  torques = re.findall(r"[0-9]{4,7}", torque)
  if len(torques) == 0:
    return 2500
  return int(torques[-1:][0])
```

```python
check = df['torque'].apply(extractRPM)
df['rpm'] = check
```

```python
df
```

Time to extract mileage,engine and max_power	

```python
import re
def extractDeci(text):
  nums = re.findall("\d+\.\d+",text)
  return float(nums[0])

def extractNum(text):
  nums = re.findall("[0-9]+",text)
  return int(nums[0])

def extractBoth(text):
  floats = re.findall("\d+\.\d+",text)
  if(len(floats) > 0):
    return float(floats[0])
  nums = re.findall("[0-9]+",text)
  return int(nums[0])
  

  
```

```python
df['mileage_num'] = df['mileage'].apply(extractDeci)
```

```python
df['engine_num'] = df['engine'].apply(extractNum)
```

```python
df['max_power'] = df['max_power'].apply(extractBoth)
```

Handling Categorical Variables

I am going to use one hot enconding for categorical features

But first we are going to drop the model column

```python
df = df.drop('model',axis=1)
```

```python
df.columns
```

```python
cat_vars = ['fuel', 'seller_type','transmission', 'owner','company']
for col in cat_vars:
  print(f"{col} , unique values => {len(df[col].unique())} ")
```

I am going to use pandas get dummies for ohe



```python
fuel_ohe = pd.get_dummies(df['fuel'],drop_first=True)
seller_type_ohe = pd.get_dummies(df['seller_type'],drop_first=True)
transmission_ohe = pd.get_dummies(df['transmission'],drop_first=True)
owner_ohe = pd.get_dummies(df['owner'],drop_first=True)
```

since company has 31 unique values we are going to use top 9 frequent values for ohe and skip other values 

```python
top_company = df['company'].value_counts()[:9].index
```

```python
def get_company(company):
  if company in top_company:
    return company
  else:
    return "Other"
```

```python
df['company_top'] = df['company'].apply(get_company)
df
```

```python
company_ohe = pd.get_dummies(df['company_top'],drop_first=True)
company_ohe
```

```python
df_final = pd.concat([df,fuel_ohe,seller_type_ohe,transmission_ohe,owner_ohe,company_ohe],axis=1)
df_final = df_final.drop(cat_vars,axis=1)
df_final = df_final.drop("company_top",axis=1)
```

```python
df_final
```

```python
X = df_final.drop("selling_price",axis=1)
y = df_final["selling_price"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### Machine learning Time

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_scaled_train,y_train)
print(f"test: {lr.score(X_scaled_test,y_test)} train : {lr.score(X_scaled_train,y_train)}")
```

**SVR**

```python
from sklearn import svm
svr = svm.SVR(kernel="linear")
svr.fit(X_scaled_train,y_train)
print(f"test: {svr.score(X_scaled_test,y_test)} train : {svr.score(X_scaled_train,y_train)}")
```

**Decision Tree Regressor**

```python
from sklearn import tree
dtr = tree.DecisionTreeRegressor()
dtr.fit(X_scaled_train,y_train)
print(f"test: {dtr.score(X_scaled_test,y_test)} train : {dtr.score(X_scaled_train,y_train)}")
```

**AdaBoost**

```python
from sklearn.ensemble import AdaBoostRegressor
regr = AdaBoostRegressor()
regr.fit(X_scaled_train,y_train)
print(f"test: {regr.score(X_scaled_test,y_test)} train : {regr.score(X_scaled_train,y_train)}")
```

**RandomForestRegreesor**

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(X_scaled_train,y_train)
print(f"test: {rf_reg.score(X_scaled_test,y_test)} train : {rf_reg.score(X_scaled_train,y_train)}")
```
